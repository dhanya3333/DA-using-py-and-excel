1.WAP TO DEMONSTARTE HOW TO DRAW A BAR PLOT USING MAT PLOT LIB


```python
import matplotlib.pyplot as plt
courses = ['Python','Java','C++','Javascript','AI']
students = [50,40,30,60,45]
plt.bar(courses,students)
plt.title('Number of Students Enrolled in Each Course')
plt.xlabel('Courses')
plt.ylabel('students')
plt.show()
```

2.WAP TO DEMONSTRATE HOW TO DRAW A SCATTER PLOT USING MAT PLOT LIB


```python
import matplotlib.pyplot as plt
x = [10,20,30,40,50]
y = [5,15,25,10,30]
plt.scatter(x,y)
plt.title("Sample Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

3.WAP TO DEMONSTRATE HOW TO DRAW A HISTOGRAM PLOT USING MAT PLOT LIB

```python
import matplotlib.pyplot as plt
ages = [12,23,45,22,36,27,29,40,50,30,28,33,31,18,21,25]
plt.hist(ages,bins=5)
plt.title("Age Distribution Histogram")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.show()
```

4.WAP TO DEMONSTRATE HOW TO DRAW A PIE CHART USING MAT PLOT LIB

```python
import matplotlib.pyplot as plt
courses = ['Python','Java','C++','AI']
students = [40,25,20,15]
plt.pie(students, labels=courses, autopct='%1.1f%%')
plt.title("Student Enrollment in Different Courses")
plt.show()
```

5.WAP TO ILLUSTRATE LINEAR PLOTTING USING MATPLOTLIB


```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.title("Simple Linear Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
